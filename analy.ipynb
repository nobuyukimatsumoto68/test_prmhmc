```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy
import math
from scipy import special as sp
```

```python
results = np.loadtxt( "results2.dat" )
```

```python
op=0
```

```python
switch=1
retrieved = results[ results[:,-1]==switch ]
tmp = np.array( [retrieved.T[4][op::2], retrieved.T[0][op::2], retrieved.T[1][op::2]] )
prmhmc = np.sort( tmp.T, axis=0 )

tmp = np.array( [retrieved.T[4][op::2], retrieved.T[2][op::2]] )
exact = np.sort( tmp.T, axis=0 )

switch=3
retrieved = results[ results[:,-1]==switch ]
tmp = np.array( [retrieved.T[4][op::2], retrieved.T[0][op::2], retrieved.T[1][op::2]] )
rmhmc_fixedmass = np.sort( tmp.T, axis=0 )

switch=4
retrieved = results[ results[:,-1]==switch ]
tmp = np.array( [retrieved.T[4][op::2], retrieved.T[0][op::2], retrieved.T[1][op::2]] )
rmhmc_variedmass = np.sort( tmp.T, axis=0 )
```

```python
plt.errorbar( prmhmc.T[0], prmhmc.T[1], prmhmc.T[2], 
             marker='x', capsize=2, label="PRMHMC")
plt.errorbar( rmhmc_fixedmass.T[0], rmhmc_fixedmass.T[1], rmhmc_fixedmass.T[2], marker='.', capsize=2, label="RMHMC (fixed mass)")
plt.errorbar( rmhmc_variedmass.T[0], rmhmc_variedmass.T[1], rmhmc_variedmass.T[2], marker='+', capsize=2, label="RMHMC (varying mass)")

plt.plot( exact.T[0], exact.T[1], label="exact", ls='dashed')

plt.legend()
```

```python
plt.errorbar( prmhmc.T[0], prmhmc.T[1]-exact.T[1], prmhmc.T[2], 
             marker='x', capsize=2, label="PRMHMC")
plt.errorbar( rmhmc_fixedmass.T[0], rmhmc_fixedmass.T[1]-exact.T[1], rmhmc_fixedmass.T[2], marker='.', capsize=2, label="RMHMC (fixed mass)")

plt.legend()
```
